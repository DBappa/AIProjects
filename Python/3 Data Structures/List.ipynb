#### List Comprehension

Basics Syntax  [expression for item in iterable]

with conditional logic [expression for item in iterable if condition]

Nested List Comprehension [expression for item1 in iterable1 for item2 in iterable2]

```python
## Basic List Comprehension

square= [x**2 for x in range(10)]
print(square)
```

```python
## List Comprehension with Condition

square= [x**2 for x in range(10) if x%2==0]
print(square)
```

```python
## Nested List Comprehension

matrix = [[1,2,3],[4,5,6],[7,8,9]]
result = [[row[i] for row in matrix] for i in range(3)]
print(result)
```

```python
## List Comprehension with function calls

words = ["hello","world","python","programming","list", "comprehension"]
result = [word.upper() for word in words]
print(result)
```
