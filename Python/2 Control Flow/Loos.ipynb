```python
## while loop

## The while loop continues to run while the condition is true
i=1
while i<6:
    print(i)
    i+=1
```

```python
## Loop Control Statements

## break statement
## The break statement is used to exit the loop prematurely
i=1
while i<6:
    print(i)
    if i==3:
        break
    i+=1

```

```python
## continue statement

## The continue statement skips the current iteration of the loop and continue with the next

i=0
while i<6:
    i+=1
    if i==3:
        continue
    print(i)
```

```python
## pass statement
## The pass statement does nothing. It can be used as a placeholder for future code
for i in [0, 1, 2]:
    pass
```

```python
## Nested Loops
## A loop inside a loop

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j: {j}")
        
```

```python
## Examples - Calculate the sum of first N natural numbers using a while and for loop

## Using while loop
i=1
sum=0
while i<=10:
    sum+=i
    i+=1
print(f"Using While Loop: {sum}")   # print("sum)

## Using for loop   
sum=0
for i in range(1,11):
    sum+=i
print(f"Using for Loop: {sum}")
```

```python
## Example -Prime number between 1 and 100


for i in range(2,101):
    for j in range(2,i):
        if i%j==0:
            break
    else:
        print(i)
```
