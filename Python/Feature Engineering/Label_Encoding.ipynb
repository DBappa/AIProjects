### Label Encoding
Label encoding and ordinal encoding are two techniques used to encode categorical data as numerical data

Label encoding involves assigning a unique numerical label to each category in the variable.The labels are usually assigned in alphabetical order based on the frequency of the categories.For example, if we have a categorical variable "color" with three possible values(red,green,blue), we can represent it using label encoding as follows: 

1. Red: 1
2. Green: 2
3. Blue: 3

```python
## Create a simple dataframe
import pandas as pd
df = pd.DataFrame({'color': ['red', 'green', 'blue', 'red', 'green', 'blue']})
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit_transform(df['color'])
```

```python
le.transform(['red'])
```

```python
le.transform(['blue'])
```

### Ordinal Encoding

It is used to encode categorical data that have an intrinsic order or ranking. In this technique, each category is assigned a numerical value based on its position in the order. For example, if we have a categorical variable "education level" with four possible values (high school, college, graduate, post graduate), we can represent it using ordinal encoding as follows:

1. High School: 1
2. College: 2
3. Graduate: 3
4. Post-graduate: 4

```python
### Ordinal Encoding 

from sklearn.preprocessing import OrdinalEncoder
## Create an instance of OrdinalEncoder
oe = OrdinalEncoder(categories=[['High School', 'College', 'Graduate', 'Post Graduate']])

## create a sample dataframe with an ordinal variable
df = pd.DataFrame({'degree':['High School', 'College', 'Graduate', 'Post Graduate']})
```

```python
oe.fit_transform(df[['degree']])
```
